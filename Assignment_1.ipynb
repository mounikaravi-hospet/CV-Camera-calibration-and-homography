# Converting the 3x4 projection matrix by eliminating the last column since the remaining 3x3  matrix is the product of Intrinsic matrix and rotation matrix.

```python
kp = np.array(
[
    [-45263300,   -45263300,   -32020000],
    [-9606500, -9606500,    -8584800],
    [-301117864,  -301117864, -1924174272]
])
```

# QR factorization of the 3x3 projection  matrix to split it into Intrinsic matrix and rotation matrix.

```python
q, r = np.linalg.qr(kp)
```
